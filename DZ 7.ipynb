```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#Наша функция
def f(x, a, b, c):
    return c/(1+((x-a)/b)**2)

xx = np.linspace(-3, 3, 50)
yy = f(xx, 0.5, 1, 2)

#Добавление шума
np.random.seed(1000)
noise = np.random.normal(loc = 0 , scale = 0.03, size = xx.size) 
noisy_y = yy + 10 * noise
```

```python
plt.plot (xx, yy, '--', label = 'Чистые данные')
plt.plot (xx, noisy_y, 'o', label = 'Шум')
plt.title ('Данные с шумами')
plt.legend (loc =(0.1, 0.8) )
```

```python
popt, pcov= curve_fit(f, xx, noisy_y)
```

```python
%matplotlib inline

plt.plot (xx, yy, '-r', label = 'Функция без шумов')
plt.plot (xx, noisy_y, 'o', label = 'Шумы') 
plt.plot (xx, f(xx, *popt), '-g', label = 'Оптимизация')
plt.legend (loc ='best' )
plt.title ('График с оптимизацией')
```
