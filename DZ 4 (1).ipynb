# ДЗ №4: вычисление производной (Ширалиева Айсель)

```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#наша функция
def f(x):
    return (x**11)
#численно ее дифференцируем
#одноточеченая схема
def derivative_left (x, h):
    return ((f(x+h)-f(x))/h)
def derivative_right (x, h):
    return ((f(x)-f(x-h))/h)
#двухточечная схема
def derivative_centr (x, h):
    return ((f(x+h/2)-f(x-h/2))/h)
#математическая производная
def math_derivative (x):
    return (11*x**10)
#графики в двойной логарифмической шкале
k = [10**(i/10) for i in range (-190,0)]
x = 2.0
y_left = [abs(derivative_left(x,h) - math_derivative(x)) for h in k]
y_right = [abs(derivative_right(x,h) - math_derivative(x)) for h in k]
y_centr = [abs(derivative_centr(x,h) - math_derivative(x)) for h in k]
#машинный эпсилон
eps = 1E-16 
#графики
plt.loglog(k, y_left,  'go', label = u'Односторонняя левая производная')
plt.loglog(k, y_left,  'ro', label = u'Односторонняя правая производная')
plt.ylabel ('log(y)')
plt.xlabel ('log(x)')
plt.title ('Функции')
plt.grid(True)
plt.legend()
plt.show()
```

Здесь наглядно видно, что односторонняя правая и левая производные - это одно и то же

## Сравнительные графики

```python
plt.loglog(k, y_left,  'go', label = u'Односторонняя левая производная')
plt.loglog(k, y_centr, 'bo', label = u'Центральная производная')
plt.ylabel ('log(y)')
plt.xlabel ('log(x)')
plt.title ('Функции')
plt.grid(True)
plt.legend()
plt.show()
```

Видно, что при h <= eps оба графика превращаются в константы, так как в этом случае производная будет равна 0 и модуль разности будет равен просто обычной производной в точке. Соответственно, как-то анализировать график ниже этих h не имеет смысла. Далее, видно, что центральная производная дает более точное приближение, т.к. минимум разности у нее меньше. По наклону заметно, что левая производная зависит от h, а центральная от h**2.
