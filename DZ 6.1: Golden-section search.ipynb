### Метод золотого сечения

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#наша исходная функция, на которой ищем минимум
def f (x): 
    f= - (math.cos(x))
    return f
```

```python
phi = (math.sqrt(5) + 1) / 2    
#вводим функцию для поиска минимума методом золотого сечения 
def golden_sec_search (f, a, b, eps):
    x1 = b - (b - a) / phi
    x2 = a + (b - a) / phi
    y1 = f (x1)
    y2 = f (x2)
    if abs(b - a) <= eps:
        x = (a + b) / 2
        return x
    if y1 >= y2:
        return golden_sec_search (f, x1, b, eps)
    else: 
        return golden_sec_search (f, a, x2, eps)
```

```python
#тестируем функцию
print (golden_sec_search(f, 4, 7, 1E-10))
print (golden_sec_search(f, -1, 1, 1E-10))
```

```python
#рисуем графики
xx = np.linspace(1, 7, 100)
gss = golden_sec_search(f, 4, 7, 1E-10)
plt.grid(True)
plt.plot(xx, [f (xk) for xk in xx], '-')
plt.plot(gss, f (gss), 'o')
plt.title ('Нахождение минимума функции на отрезке')
```
