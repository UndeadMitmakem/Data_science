```python
import sympy
import scipy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
```

```python
t = np.linspace(0, 30*np.pi, 1000)
def f(y, t, xi, omeg1, omeg2):
    x, v = y
    return [v, -(omeg1**2)*(1-xi*x**2/100)*x + np.cos(omeg2*t)]
result = scipy.integrate.odeint(f, [0, 0], t, args = (1, 3, 3)) 
x = result
```

```python
result.shape
```

```python
x, v = result[:, 0], result[:, 1]
```

```python
plt.plot(t, x)
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot (t, x)
ax2.plot (t, v)
ax1.grid(True)
ax2.grid(True)
plt.show()
```
