# Метод бисекции

```python
import math
k = 0.1
x_left = 0
x_right = 1.0 / k
eps = 1E-10
#наша функция
def f(x, k):
    return abs(math.sin(x)) - k*x
#функция для бисекции на отрезках монотонности
def bisection(k):
    solution = [0.0]
    right = math.pi/2.0
    left = 0.0
    while right <= 1/k:       
        a = left
        b = right      
        if f(a,k) * f(b,k) < 0: 
            c=(a + b) / 2
            while abs(f(c,k)) > eps:
                c =(a + b) / 2
                if f(c,k) * f(a,k) < 0:
                    b=c
                else:
                    a=c
            solution.append(c)         
        left += math.pi/2 
        right += math.pi/2
    return solution
#функция для выбора случаев
def general (k):
    if k == 0:
        return "pi*n"
    elif abs(k) >= 1:
        return 0.0
    elif k > 0:
        solutions = bisection(k)
        return solutions
    else:
        k = -1*k
        negative = [-1 * i for i in bisection(k)]
        solutions = negative 
        return solutions
```

Тестируем для различных случаев
